```python
def generate_video(prompt, num_frames=16, num_steps=25, height=256, width=256):
    try:
        video_frames = pipe(
            prompt,
            num_inference_steps=num_steps,
            num_frames=num_frames,
            height=height,
            width=width
        ).frames

        # Process frames to ensure correct channels
        processed_frames = process_frames(video_frames)

        # Export to video
        video_path = "generated_video.mp4"
        export_to_video(processed_frames, video_path)
        return video_path

    except Exception as e:
        print(f"Error generating video: {str(e)}")
        return None
    finally:
        torch.cuda.empty_cache()
```
